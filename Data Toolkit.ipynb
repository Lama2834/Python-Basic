2.How does broadcasting work in NumPy?

NumPy Broadcasting

- Broadcasting in NumPy is a powerful mechanism that allows operations on arrays of different shapes without the need for explicit looping or reshaping. It enables element-wise operations between arrays of different sizes by automatically expanding their shapes to make them compatible.

- How Broadcasting Works

When performing operations between two arrays, NumPy compares their shapes element-wise. The arrays are broadcasted (expanded) following these rules:

If the dimensions of the arrays are different, the smaller array is left-padded with ones until both arrays have the same number of dimensions.
NumPy compares dimensions element-wise:
If they are equal, they are compatible.
If one dimension is 1, it is stretched to match the other dimension.
If the dimensions differ and neither is 1, an error occurs.
The operation is performed element-wise on the expanded arrays.

4. A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample dataset
df = sns.load_dataset("penguins").dropna()  # Load dataset and remove NaN values

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))  # Set figure size
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```

8. Create a line plot with multiple lines using Matplotliba.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y1 = np.sin(x)  # First line (Sine wave)
y2 = np.cos(x)  # Second line (Cosine wave)
y3 = np.tan(x)  # Third line (Tangent wave)

# Create the plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y1, label="Sine Wave", color="blue", linestyle="-")
plt.plot(x, y2, label="Cosine Wave", color="red", linestyle="--")
plt.plot(x, y3, label="Tangent Wave", color="green", linestyle=":")

# Add labels, title, and legend
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Multiple Line Plot Example")
plt.legend()
plt.ylim(-2, 2)  # Set y-axis limits to avoid extreme tangent values

# Show the plot
plt.show()
```

13.Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(42)
df = pd.DataFrame({
    "X": np.random.randn(100),
    "Y": np.random.randn(100),
    "Z": np.random.randn(100),
    "Category": np.random.choice(["A", "B", "C"], 100)  # Assign categories
})

# Create 3D scatter plot
fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Category",
                     title="Interactive 3D Scatter Plot", size_max=10)

# Show the plot
fig.show()
```
